```python
# The metrics scores we are using are from the model evaluation
def add_value_label(x_list,y_list):
    for i in range(1, len(x_list)+1):
        plt.text(i-1.2,y_list[i-1]+1,y_list[i-1])
        
graph_data = {'metrics': ['Accuracy', 'Precision', 'Recall', 'F1-score'],
             'scores': [97.78, 96.97, 97.62, 97.18]} 

df = pd.DataFrame(graph_data, columns=['metrics', 'scores'])

colors = ['teal', 'purple', 'brown', 'green']
fig = plt.figure()
plt.bar(df['metrics'], df['scores'], color=colors, width=0.5, align='center')
add_value_label(df['metrics'], df['scores'])
plt.title('Model Evaluation')
plt.xlabel('Evaluation Metrics')
plt.ylabel('Score %')
# plt.text(84, 95, '93.33%')
plt.show()
fig.savefig("model_evaluation2")
```

```python
# 
def add_value_label(x_list,y_list):
    for i in range(1, len(x_list)+1):
        plt.text(i-1.2,y_list[i-1]+1,y_list[i-1])
        
graph_data = {'cat': ['Blockbuster', 'Hit', 'Flop'],
             'values': [99, 65, 59]} 

df = pd.DataFrame(graph_data, columns=['cat', 'values'])

colors = ['green', 'teal', 'brown']
fig = plt.figure()
plt.bar(df['cat'], df['values'], color=colors, width=0.5, align='center')
add_value_label(df['cat'], df['values'])
plt.title('Category count')
plt.xlabel('Categories')
plt.ylabel('Count')

plt.show()
fig.savefig("cat_count")
```
