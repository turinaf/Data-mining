```python
import pandas as pd
```

```python
dataset = pd.read_csv('final_movie_dataset.csv')
dataset.head()
```

### Actors and Director Rating

Now we want to give rating for main actor, second actor and Director, out of 10 based on number of movies he/shi has made and out of those movies which are Blockbuster, Hit or Flip.
The score for category of a movie an actor made is as follows.
>* BLOCKBUSTER = 10
>* HIT = 7.5
>* FLOP = 5
###### The rating for the main actor is as follows.
Main_actor_rate = [(# of blockbuster movies the actor made * BLOCKBUSTER) + (# of Hit movies the actor made * HIT) + (# of Flop movies the actor made * FLOP)] divided by (BLOCKBUSTER + HIT + FLOP)

#### 1. Main actor rating

```python
Blockbuster = []
Hit = []
Flop = []

for i in range(len(dataset)):
    blockbuster = 0
    hit = 0
    flop = 0
    for j in range(len(dataset)):
        b = (dataset.Main_actor[i] == dataset.Main_actor[j])
        c = dataset.Category[j] == 'Blockbuster'
        d = dataset.Category[j] == 'Hit'
        e = dataset.Category[j] == 'Flop'
        if (b and c) is True:
            blockbuster = blockbuster+1
        if (b and d) is True:
            hit = hit+1
        if (b and e) is True:
            flop = flop+1
    Blockbuster.append(blockbuster)
    Hit.append(hit)
    Flop.append(flop)
    
dataset['Blockbuster'] = Blockbuster
dataset['Hit'] = Hit
dataset['Flop'] = Flop

dataset['Main_actor_rate'] = ((dataset['Blockbuster']*10) + 
                             (dataset['Hit']*7.5) + 
                             (dataset['Flop']*5))/ (dataset['Blockbuster'] + dataset['Hit'] + dataset['Flop'])

dataset.drop(['Blockbuster', 'Hit', 'Flop'], inplace=True, axis=1)
```

#### 2. Director Rating
The score for the three categories of movies are same as we did for Main character rating above.

```python
Blockbuster = []
Hit = []
Flop = []

for i in range(len(dataset)):
    blockbuster = 0
    hit = 0
    flop = 0
    for j in range(len(dataset)):
        b = (dataset.Director[i] == dataset.Director[j])
        c = dataset.Category[j] == 'Blockbuster'
        d = dataset.Category[j] == 'Hit'
        e = dataset.Category[j] == 'Flop'
        if (b and c) is True:
            blockbuster = blockbuster+1
        if (b and d) is True:
            hit = hit+1
        if (b and e) is True:
            flop = flop+1
    Blockbuster.append(blockbuster)
    Hit.append(hit)
    Flop.append(flop)
    
dataset['Blockbuster'] = Blockbuster
dataset['Hit'] = Hit
dataset['Flop'] = Flop

dataset['Director_rate'] = ((dataset['Blockbuster']*10) + 
                             (dataset['Hit']*7.5) + 
                             (dataset['Flop']*5))/ (dataset['Blockbuster'] + dataset['Hit'] + dataset['Flop'])

dataset.drop(['Blockbuster', 'Hit', 'Flop'], inplace=True, axis=1)

dataset.head()
```

#### 2. Second Actor Rating
The score for the three categories of movies are same as we did for Main character rating above.

```python
Blockbuster = []
Hit = []
Flop = []

for i in range(len(dataset)):
    blockbuster = 0
    hit = 0
    flop = 0
    for j in range(len(dataset)):
        b = (dataset.Second_actor[i] == dataset.Second_actor[j])
        c = dataset.Category[j] == 'Blockbuster'
        d = dataset.Category[j] == 'Hit'
        e = dataset.Category[j] == 'Flop'
        if (b and c) is True:
            blockbuster = blockbuster+1
        if (b and d) is True:
            hit = hit+1
        if (b and e) is True:
            flop = flop+1
    Blockbuster.append(blockbuster)
    Hit.append(hit)
    Flop.append(flop)
    
dataset['Blockbuster'] = Blockbuster
dataset['Hit'] = Hit
dataset['Flop'] = Flop

dataset['Second_actor_rate'] = ((dataset['Blockbuster']*10) + 
                             (dataset['Hit']*7.5) + 
                             (dataset['Flop']*5))/ (dataset['Blockbuster'] + dataset['Hit'] + dataset['Flop'])

dataset.drop(['Blockbuster', 'Hit', 'Flop'], inplace=True, axis=1)

dataset.head()
```

### Adjusting metascore and rotten_tomatoes
Since this two column are in percent (%), we have to do processing to adjust it.


```python
rottent = dataset['rotten_tomatoes']
metascore = dataset['metascore']
for j in range(len(rottent)):
    rottent[j] = int(rottent[j])/10.0
    metascore[j] = int(metascore[j])/10.0
print(rottent)
print(metascore)
```
